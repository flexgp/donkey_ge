# Run Program Synthesis

```python
data_file = "tests/program_synthesis/FindRange.json"
with open(data_file, "r") as f:
    data = json.load(f)

program_synthesis = FindRange(data["train"], "")
code = """
    res0 = 0
    for i in inputs:
        if i >= "b" and i <= "f":
            res0 = res0 + 1
    return res0
    """
result = program_synthesis.run(code)
print(result)
assert (sum(result) == 100, result)
```
