# Run Program Synthesis

```python
data_file = "tests/program_synthesis/ParenthesisMatching.json"
with open(data_file, "r") as f:
    data = json.load(f)
```

```python
program_synthesis = ParenthesisMatching(data["train"], "")
code = """
        self.increment = 1
        self.decrement = 1
        self.check = 0
        left = 0
        for i in inputs:
            if i == "(":
                left += self.increment
            elif left == self.check:
                return False
            else:
                left -= self.decrement
        return left == self.check
    """
result = program_synthesis.run(code)
print(result)
assert sum(result) == 500
```
